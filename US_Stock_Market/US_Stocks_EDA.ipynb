# Volatility

```python
list = ["Date"]
for i in df.columns[1:]:
    if 'Vol' not in i:
        list.append(i)
list
```

```python
columns_to_keep = list
df_price = df[columns_to_keep].copy()
```

```python
df_price.info()
```

```python
def convert_to_float(col):
    df_price[col] = df_price[col].str.replace(',', '').astype(float)

for col in df_price.columns[1:]:  
    if df_price[col].dtype == 'object':
        convert_to_float(col)
```

```python
df_price
```

```python
df_volatile = df_price.columns[1:]
df_volatile = pd.DataFrame(df_volatile)
df_volatile.rename(columns = {0 : "Stocks"}, inplace = True)
```

```python
df_volatile
```

```python
def Volatility(col):
    returns_df = df_price[col].pct_change()
    volatility = returns_df.std()
    return round((volatility*100),2)
```

```python
df_volatile["Volatility in (%)"] = df_volatile["Stocks"].apply(Volatility)
```

```python
#df_volatile.sort_values(by = "")
```

```python
import seaborn as sns
sns.barplot(x = 'Stocks', y = 'Volatility in (%)', data = df_volatile)
plt.title("Volatility of Socks")
plt.xticks(rotation = 90)
plt.show()
```

```python
df_volatile.sort_values(by = 'Volatility in (%)', ascending = False).set_index("Stocks")
```

# Stocks with highest returns 

LET US CALCULATE THE STOCK THAT GAVE HIGHEST RETURN.. FROM 2020 -2024

```python
df_return = df_price.columns[1:]
df_return = pd.DataFrame(df_return)
df_return
```

```python
df_return.rename(columns = {0 : "Stocks"}, inplace = True)
```

```python
df_return
```

LET ME GO WITH STRAIGHT FORWARD METHOD OF FINDING THE RETURN

```python
def cal_return(col):
    return round((((df_price[col][1012] - df_price[col][0])/df_price[col][0])*100),2)
```

```python
df_return["returns in %"] = df_return["Stocks"].apply(cal_return)
```

```python
df_return
```

```python
sns.barplot(x = "Stocks", y = "returns in %", data = df_return)
plt.xticks(rotation = 90)
plt.grid(True, color='gray', linestyle='--', linewidth=0.5)
plt.title("returns by stocks")
plt.show()
```

```python
df_return.sort_values(by = "returns in %", ascending = False).set_index("Stocks")
```
